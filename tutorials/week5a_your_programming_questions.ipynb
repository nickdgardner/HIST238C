## 1. Python question: on "control flow" (while loops, for loops, if/elif/else)

### While loops

The code "within the scope" of a while loop (i.e. all the code that is indented inward at least once, relative to the ``while`` keyword) gets executed ("run") repeatedly as long as while loops's condition is true. Examples of while loop conditions in the cells below are:

``while i <= 10:  #this means "as long as i is less than or equal to 10"
 while i < 2:    #this means "as long as i is strictly less than 2"``


```python
#print the first ten positive integers
#the two lines of code within the while loop's scope will be executed exactly ten times
i = 1
while i <= 10: #"as long as the variable i is less than or equal to 10, execute the code below"
    print(i)
    i += 1  #increment i by one -- "i += 1" is shorthand for "i = i + 1"

#
print("the value of i is now " + str(i)) #i ends up being 11, which is why the while loop terminates
```

```python
#we can nest as many while loops within one another as we want.

#before executing this cell, try to think what the output of the code below will be
#(a good way to do this is to write it out on paper, keeping track of how the variables
#i and j are updated

i = 0
while i < 2:
    j = 0
    while j < 3:
        print(i, j)
        j += 1
    i += 1
```

```python
#an example from Automate the Boring stuff
i = 0
while i < 5:
    print("Jimmy Five Times (" + str(i) + ")") 
    i = i + 1
```

### For loops

```python
#the code in this cell will have same result as while loop above

#create a list of integers and name it something informative, for example "iterations"
iterations = [0,1,2,3,4] 

#for ITEM in LIST is the syntax for iterating through each item in a list
for i in iterations: 
    print("Jimmy Five Times (" + str(i) + ")") 
```

```python
#the standard/idiomatic way to do the above, using the range() function
#range(5) returns something similar to the list [0,1,2,3,4] above
for i in range(5):
    print("Jimmy Five Times (" + str(i) + ")") 
```

### Conditionals
`if`, `elif` (short for "else if"), `else`

```python
#to illustrate what if, elif, and else do, we will iterate through a (made up) corrupted
#sequence of DNA that has some "noise" in it (junk symbols--imagine this as resulting from
# random errors in measurement). We will count
#how many times the DNA base symbols A, C, T, G appear, and 
#also count how many junk symbols appear (every symbol that is not A, C, T, or G)

#initialize a string variable to a DNA sequence with some random corruptions/noise
corrupted_DNA_seq = "ACAAGATGCq3CATTGiiip2TCCCCCGfwGCCTcCCTGzCTGCTGCTweGCTCTCCGGGGCqCACGGsCCAxCCGCTGCCCTGCC"


#initialize a dict(ionary) of five counters 
#the keys of the dict are the symbols
#the value for each key is the number of times the symbol appears
counter_dict = {'A':0, 
               'C':0,
               'T':0,
               'G':0,
               'junk':0}


#use a loop to count base symbols 'A' 'C' 'T' 'G', and junk symbols
for character in corrupted_DNA_seq: #note we are iterating through the characters in a string--strings are like lists!
    if character == 'A': #note the difference between "=" (variable assignment operator) and "==" (test if two expressions are equal)
        counter_dict['A'] += 1 #shorthand for incrementing a variable by 1. Same as counter_dict['A'] = counter_dict['A'] + 1
    
    elif character == 'C':  #this elif check will get executed only if the current character is not 'A' 
        counter_dict['C'] += 1
    
    elif character == 'T': #this elif check will get executed only if the current character is not 'A' or 'C'
        counter_dict['T'] += 1
    
    elif character == 'G': #this elif check will get executed only if the current character is not 'A', 'C', or 'T'
        counter_dict['G'] += 1
    
    else: #this elif check will get executed only if the current character was not 'A', 'C','T', or 'G'--ie it is a junk symbol
        counter_dict['junk'] += 1
```

## 2. Python question: what does % do in a string? 
`%` in a string is called the "string formatting operator". The point of the `%` operator is to be able to plug the value of a variable into a certain place of a string you want to print. Within the string, the `%` has to be followed by a letter that indicates the type of the variable you want to plug in. `%d` means "I'm going to plug in a number here". `%s` means "I'm going to plug in a string here".

Don't worry about understanding this syntax very thoroughly, which is a legacy of the C programming language. In python there are more intuitive ways to do this (see the last cell below).

```python
wins = 5
losses = 6
ties = 2
print('%d Wins, %d Losses, %d Ties' % (wins, losses, ties))
```

```python
#if we change the values of the variables, this will be reflected in the string we print out
wins = 0
losses = 0
ties = 0
print('%d Wins, %d Losses, %d Ties' % (wins, losses, ties))
```

```python
#an example plugging in string variables
dept1 = "History"
dept2 = "English" #error -- should be English
dept3 = "Computer Science"
print('This class combines ideas from three departments: %s, %s, and %s' % (dept1, dept2, dept3))
```

```python
#an easier way to do the above, using the string method .format(). The variables get plugged in to where the brackets
#are, in order
print('{} Wins, {} Losses, {} Ties'.format(wins, losses, ties))
print('This class combines ideas from three departments: {}, {}, {}'.format(dept1, dept2, dept3))

#with .format() we can plug in different data types without having to worry about specifying them within
#the string
print('Plug in a number here: {}. And plug in a string here: {}'.format(5, "hello!"))
```

## 3. Pandas question: how can I calculate duration of time spent in Florence for all travelers who went there



```python
#get the subset of itinerary rows corresponding to a trip to florence
florence_travelers = travelers_itineraries_all[ travelers_itineraries_all['travelPlace'] == 'Florence' ]
```

```python
#inspect it to make sure what we did seems right
florence_travelers.head(20)
```

```python
#pd.to_datetime is a really cool function that converts format to "datetime" objects that you can do math on (e.g.
#subtract the date 1757-5-01 from the later date 1759-4-9, to get the number of days between these two dates.
#So, you don't have to worry about different months having different numbers of days, or leap years
start_dates = pd.to_datetime(florence_travelers['startDate'])
end_dates = pd.to_datetime(florence_travelers['endDate'])
```

```python
florence_trip_durations = end_dates - start_dates
```

```python
#note this data would need to be cleaned and inspected (suspicious travel durations include 0 and 365 days, and
#extremely long durations)
florence_trip_durations.head(50)
```

```python
#alternatively, you could add start_dt, end_dt, and trip_duration columns to the itineraries dataframe itself
#for now don't worry if you get a warning from Jupyter
florence_travelers['start_dt'] = start_dates
florence_travelers['died_dt'] = end_dates
florence_travelers['trip_duration'] = end_dates - start_dates

#see righmost three columns
florence_travelers.head(20)
```
